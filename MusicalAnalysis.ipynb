```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

path='/Users/avernellchristopher/Documents/Project7/Songs'
extension = '.csv'


```

```python
files = [file for file in os.listdir(path) if file.endswith(extension)]
dfs =[]
for file in files:
    df= pd.read_csv(os.path.join(path,file))
    dfs.append(df)
    
all_top_songs=pd.concat(dfs,ignore_index=False)



```
